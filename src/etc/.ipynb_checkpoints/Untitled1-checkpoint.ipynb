```python
gts=open("/media/syh/ssd2/SynologyDrive/DB/인증시험용_데이터/gt.txt",'r')

gt_dict={}

for i in range(100):
    gt_dict[str(i+1)]=[]

while True:
    line = gts.readline()
    if not line: break
    
    f_num,f_id,x,y,w,h = line.split(" ")

    gt_dict[str(f_num)].append([int(f_id),float(x),float(y),float(w),float(h[:-1])])

gts.close()
```
