```python
from PIL import Image
import os

def resize_image(image_path):
    with Image.open(image_path) as img:
        img_resized = img.resize((960, 540))
    return img_resized

def merge_images(a_path, b_path, c_path, d_path, output_path):
    # 이미지 열기 및 크기 조절
    a = resize_image(a_path)
    b = resize_image(b_path)
    c = resize_image(c_path)
    d = resize_image(d_path)

    # 새로운 이미지 생성 (1920x1080)
    # new_img = Image.new('RGB', (1920, 1080))
    new_img = Image.new('BGR', (1920, 1080))

    # 이미지 붙이기
    new_img.paste(a, (0, 0))
    new_img.paste(b, (960, 0))
    new_img.paste(c, (0, 540))
    new_img.paste(d, (960, 540))

    # 이미지 저장
    new_img.save(output_path)

# 폴더 경로
a_folder = join_1_img_path
b_folder = join_2_img_path
c_folder = gjtp_1_img_path
d_folder = gjtp_2_img_path
output_folder = 'Output'

# 각 폴더에서 이미지 경로 가져오기
a_images = sorted([os.path.join(a_folder, fname) for fname in os.listdir(a_folder)])
b_images = sorted([os.path.join(b_folder, fname) for fname in os.listdir(b_folder)])
c_images = sorted([os.path.join(c_folder, fname) for fname in os.listdir(c_folder)])
d_images = sorted([os.path.join(d_folder, fname) for fname in os.listdir(d_folder)])

# 이미지 합치기
for i in range(100):
    output_path = os.path.join(output_folder, (str(i)+'.jpg').zfill(10))
    merge_images(a_images[i], b_images[i], c_images[i], d_images[i], output_path)


```

```python
'''
jointree -> tlbr
gjtp -> xywh

All into xywh

230727/
0 101 0.579167 0.317130 0.100000 0.399074
0 102 0.782031 0.649537 0.110937 0.493519
0 103 0.644010 0.200000 0.068229 0.344444
0 104 0.867448 0.360648 0.094271 0.391667
0 105 0.717188 0.153704 0.075000 0.305556
0 106 0.959635 0.265741 0.080729 0.342593
'''
# 이미지 합치기 및 txt 파일 합치기 코드
def merge_txt_files(a_path, b_path, c_path, d_path, output_path,ind):
    with open(a_path, 'r') as f:
        a_data = [update_coordinates_tlbr(line, 0, 0) for line in f.readlines()]

    with open(b_path, 'r') as f:
        b_data = [update_coordinates_tlbr(line, 0.5, 0) for line in f.readlines()]

    with open(c_path, 'r') as f:
        c_data = [update_coordinates_xywh(line, 0, 0.5) for line in f.readlines()]

    with open(d_path, 'r') as f:
        d_data = [update_coordinates_xywh(line, 0.5, 0.5) for line in f.readlines()]
    with open(output_path, 'w') as f:
        f.writelines(c_data+a_data + d_data + b_data )
        
    with open(a_path, 'r') as f:
        a_data = [update_coordinates_tlbr(line, 0, 0,ind) for line in f.readlines()]

    with open(b_path, 'r') as f:
        b_data = [update_coordinates_tlbr(line, 0.5, 0,ind) for line in f.readlines()]

    with open(c_path, 'r') as f:
        c_data = [update_coordinates_xywh(line, 0, 0.5,ind) for line in f.readlines()]

    with open(d_path, 'r') as f:
        d_data = [update_coordinates_xywh(line, 0.5, 0.5,ind) for line in f.readlines()]    
    
    return c_data+a_data + d_data + b_data 



def update_coordinates_xywh(line, dx, dy,ind=None):
    parts = line.strip().split(' ')
    x = float(parts[2])/2 + dx
    y = float(parts[3])/2 + dy
    w = float(parts[4])/2
    h = float(parts[5])/2
    if ind == None:
        updated_line = f"{parts[0]},{parts[1]},{x:.6f},{y:.6f},{w:.6f},{h:.6f}\n"
    else:
        updated_line = f"{ind},{parts[1]},{x:.6f},{y:.6f},{w:.6f},{h:.6f}\n"
    return updated_line

def update_coordinates_tlbr(line, dx, dy,ind=None):
    parts = line.strip().split(' ')
    w = float(parts[4]) - float(parts[2]) # w
    h = float(parts[5]) - float(parts[3]) # w
    parts[2] = float(parts[2]) + w/2 # cx
    parts[3] = float(parts[3]) + h/2 # cy
    
    x = parts[2]/2 + dx
    y = parts[3]/2 + dy
    w = w/2
    h = h/2



    # x1 = float(parts[2])/2 + dx
    # y1 = float(parts[3])/2 + dy
    # x2 = float(parts[4])/2 + dx
    # y2 = float(parts[5])/2 + dy
    # w = x2-x1
    # h = y2-y1
    if ind == None:
        # updated_line = f"{parts[0]},{parts[1]},{x1:.6f},{y1:.6f},{w:.6f},{h:.6f}\n"
        updated_line = f"{parts[0]},{parts[1]},{x:.6f},{y:.6f},{w:.6f},{h:.6f}\n"
        # updated_line = f"{parts[0]},{parts[1]},{x1:.6f},{y1:.6f},{w:.6f},{h:.6f}\n"
    else:
        # updated_line = f"{ind},{parts[1]},{x1:.6f},{y1:.6f},{w:.6f},{h:.6f}\n"
        updated_line = f"{ind},{parts[1]},{x:.6f},{y:.6f},{w:.6f},{h:.6f}\n"
        # updated_line = f"{ind},{parts[1]},{x1:.6f},{y1:.6f},{w:.6f},{h:.6f}\n"
    return updated_line

# 폴더 경로
a_folder = join_1_gt_path
b_folder = join_2_gt_path
c_folder = gjtp_1_gt_path
d_folder = gjtp_2_gt_path
output_folder = 'Output_label'

# 각 폴더에서 txt 파일 경로 가져오기
a_txts = sorted([os.path.join(a_folder, fname) for fname in os.listdir(a_folder) if fname.endswith('.txt')])
b_txts = sorted([os.path.join(b_folder, fname) for fname in os.listdir(b_folder) if fname.endswith('.txt')])
c_txts = sorted([os.path.join(c_folder, fname) for fname in os.listdir(c_folder) if fname.endswith('.txt')])
d_txts = sorted([os.path.join(d_folder, fname) for fname in os.listdir(d_folder) if fname.endswith('.txt')])

# txt 파일 합치기
merges=[]
for i in range(100):
    output_path = os.path.join(output_folder, (str(i)+'.txt').zfill(10))
    merges+=merge_txt_files(a_txts[i], b_txts[i], c_txts[i], d_txts[i], output_path,i)
    with open('gt.txt', 'w') as f:
        f.writelines(merges)
```
